```python
import pandas as pd

# Load CSV data
csv_data = pd.read_csv("C:\\Users\\nithy\Downloads\\supermarket_sales - Sheet1.csv")
csv_data.head(5)
```

```python
csv_data.drop(columns=['Tax 5%','Time','cogs', 'gross margin percentage', 'gross income', 'Rating'], inplace=True)
csv_data.head(5)
```

```python
csv_data.shape
```

```python
# Load JSON data
json_data = pd.read_json("C:\\Users\\nithy\Downloads\\supermarket_sales - Sheet1.json")
json_data.head()
```

```python
json_data.drop(columns=['Time','Tax 5%','cogs', 'gross margin percentage', 'gross income', 'Rating'], inplace=True)
json_data.head(5)
```

```python
json_data.shape
```

```python
excel_data = pd.read_excel("C:\\Users\\nithy\\OneDrive\\sale.xlsx")
excel_data.head()
```

```python
excel_data.shape
```

```python
#
csv_data.fillna('N/A', inplace=True)
json_data.fillna('N/A', inplace=True)
excel_data.fillna('N/A', inplace=True)
```

```python

csv_data.drop_duplicates(inplace=True)
json_data.drop_duplicates(inplace=True)
excel_data.drop_duplicates(inplace=True)
```

```python

csv_data.columns = map(str.lower, csv_data.columns)
json_data.columns = map(str.lower, json_data.columns)
excel_data.columns = map(str.lower, excel_data.columns)
```

```python
# Convert date columns to datetime
csv_data['date'] = pd.to_datetime(csv_data['date'])
json_data['date'] = pd.to_datetime(json_data['date'])
excel_data['date'] = pd.to_datetime(excel_data['date'])
```

```python

unified_data = pd.concat([csv_data, json_data, excel_data], ignore_index=True)
```

```python

print(unified_data.columns)


unified_data.columns = unified_data.columns.str.strip()


print(unified_data.columns)


print(unified_data.head())
```

```python

if 'sales' not in unified_data.columns:
    
    unified_data['sales'] = float('nan')  


unified_data['sales'] = unified_data['sales'].astype(float)
```

```python
# Concatenate all dataframes into a single DataFrame
df_combined = pd.concat([csv_data,json_data,excel_data], ignore_index=True)

# Reset the index after concatenation
df_combined.reset_index(drop=True, inplace=True)

# Display the first few rows of the combined DataFrame
df_combined.head()
```
